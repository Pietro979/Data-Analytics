```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
data = pd.read_csv("Data1.csv",index_col=0) 
```

```python
data.head()
```

```python
data.plot(y='theta_1', ylim=[-5,5])
plt.grid()
```

```python
fig, axs = plt.subplots(2,3)
data.plot(kind = 'line', y='theta_1',ax=axs[0,0], ylim=[-10, 20])
data.plot(kind = 'line', y='theta_2',ax=axs[0,1], ylim=[-10, 20])
data.plot(kind = 'line', y='theta_3',ax=axs[0,2], ylim=[-10, 20])
data.plot(kind = 'line', y='theta_4',ax=axs[1,0], ylim=[-10, 20])
data.plot(kind = 'line', y='theta_5',ax=axs[1,1], ylim=[-10, 20])
data.plot(kind = 'line', y='theta_6',ax=axs[1,2], ylim=[-10, 20])
```

```python
data.plot()
```

```python
data.head()
```

```python
dtMelt=pd.melt(data.reset_index(), id_vars = 'index', value_vars=data.columns)
dtMelt.head()
```

```python
g=sns.FacetGrid(dtMelt, col='variable', sharex=False)
g=g.map(plt.hist,"value",bins=64)
```

```python
sns.distplot(data['theta_1'],hist=True)
```

```python
data_slice = data.loc['2018-01-01':'2018-12-31','theta_1':'theta_4']
data_slice.head()
f, axes = plt.subplots(2, 2, figsize=(7, 7), sharex=True)
sns.despine(left=True)
sns.distplot(data_slice['theta_1'],hist=True, ax=axes[0, 0])
sns.distplot(data_slice['theta_2'],hist=True, ax=axes[0, 1])
sns.distplot(data_slice['theta_3'],hist=True, ax=axes[1, 0])
sns.distplot(data_slice['theta_4'],hist=True, ax=axes[1, 1])
```
